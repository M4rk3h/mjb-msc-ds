Let's consider the function
\begin{equation}
f(x) = x^4
\end{equation}

```python
# gradient descent

num_epochs = 20000
step_size = 0.1
start = 2

def fun(x):
    return x**4
    
def der(x):
    return 4*x**3

for i in range(num_epochs):
    if (i == 0):
        x = start
    x = x - step_size * der(x)
    
print(x, fun(x)) # can print at each step if in the loop.



```

```python
import numpy as np
# set our variables
tolerance = 0.0001
step_size = 0.1
x = 2
# f of x which is x^4
def fun(x):
    return x**4
# derivitive
def der(x):
    return 4*x**3
# calculate x, gradient and step_size for each step
def step(x, grad, step_size):
    return x - step_size * grad
# loop through each step, printing the x and y values
while True:
    print(x, fun(x))
    next_x = step(x, der(x), step_size)
    if (np.abs(x - next_x) < tolerance):
        break
    x = next_x
```

Let's apply the gradient descent algorithm to the function
\begin{equation}
f(x, y) = x^2 + y^2
\end{equation}

```python
# Do it for 2
tolerance = 0.01
step_size = 0.1
x = 2
y = 0.5
# f of x which is x^4
def fun(x, y):
    return x**2 + np.exp(y**2)
# derivitive x
def der_x(x, y):
    return 2*x
# derivitive y
def der_y(x, y):
    return 2*y*np.exp(y**2)
# calculate x, gradient and step_size for each step
def step(s, grad, step_size):
    return s - step_size * grad
# loop through each step, printing the x and y values
while True:
    print(x, y, fun(x,y))
    
    # gradient for x
    grad_x = der_x(x,y)
    # gradient for y
    grad_y = der_y(x,y)
    # next x    
    next_x = step(x, grad_x, step_size)
    # next y
    next_y = step(y, grad_y, step_size)
    
    if ((x - next_x)**2 + (y - next_y)**2 < tolerance**2):
        break
    x, y = next_x, next_y  
```
