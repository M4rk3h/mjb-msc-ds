# Training a single neuron neural network
The code below enables you to build and train a neural network, consisting of a single logistic neuron. It follows the structure developed in the "Logistic Regression as a Neural Network" session of the notes:

    obtain the training data
    visualise the training data
    initialise weights, bias, stepsize, num_epochs
    for 1 to num_epochs do:
        for each training example do:
            calculate the output
            calculate its contribution to error
            calculate its contribution to the training steps
        calculate overall error
        calculate training steps
        update the weights and bias by the training steps

Note: we will assume that the data has been cleaned up and preprocessed as necessary, so can be used directly to train our network.

## Setup
**Imports**

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

**Obtain and visualise data**

```python
X_train,y_train = make_blobs(n_samples=1000,centers=[(1,7),(2,0)],random_state=1)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], marker='o', c=y_train, edgecolor='k')
plt.xlabel('x0')
plt.ylabel('x1')
plt.title('Two classes to separate')
plt.axis('equal');
```

In order to produce the required Python program we are going to create 3 functions:
**logistic**, **output** and **error**. **Logistic** will provide the logistic function for us; **output** will use the logistic function and the weights, bias and inputs to calculate the neuron output; and **error** will calculate the error on a single training example.

```python
def logistic(z):
    """Calculate logistic function applied to z"""
    return 1/(1 + np.exp(-z))

x = np.linspace(-10,10,endpoint=True)
f = logistic(x)
plt.plot(x,f)
plt.title('Logistic function')
plt.xlabel('z')
plt.ylabel('f(z)');
```

```python
def output(w,b,x):
    """Calculate the output of a single logistic neuron with parameters w and b, and input x
       Note: w and x need to have the same number of columns"""
    return logistic(np.matmul(w,x.transpose()) + b)

w = np.array([1,2])
x = np.array([-0.3,0.5])
b = 0.2
print('Expected answer logistic(0.9) =', logistic(0.9))
print('Actual   answer output(w,b,x) =', output(w,b,x))
```

```python
def error(yhati, yi):
    """Calculate the error on a single example
       Note: yhati and yi need to be both scalars or both vectors of the same length"""
    return -yi*np.log(yhati) - (1-yi)*np.log(1-yhati)
```

**Gradient descent for Logistic Regression**

For a logistic regression model we can initialise the parameters to zero, without creating problems. (For neural networks with more neurons and layers we can't). We then implement the pseudocode from the notes. 

```python
m, k = X_train.shape
w = np.zeros((k))
b = 0.0
stepsize = 0.1
num_epochs = 1000
errorlog = np.zeros((num_epochs))

for i in range(num_epochs):
    E = dz = db = 0 
    dw = np.zeros(w.shape)
    for xi, yi in zip(X_train,y_train):
        yhati = output(w,b,xi)
        E  += error(yhati,yi)
        dw += (yhati - yi) * xi
        db += (yhati - yi)
    errorlog[i] = E/m
    dw /= m
    db /= m
    w -= stepsize * dw
    b -= stepsize * db
print('w = ',w,', b = ', b)
plt.plot(errorlog)
plt.title('Error behaviour during training')
plt.xlabel('Epoch')
plt.ylabel('Error');
        
        
```

How does our classifier perform? We will evaluate its performance on the training set, and on a test set. If the prediction is greater or equal to 0.5 we will predict class 1, and less than 0.5 we will predict class 0.

```python
from sklearn.metrics import confusion_matrix
y_hat_train  = output(w,b,X_train) 
y_train_pred = (y_hat_train >= 0.5).astype(int)
confusion_matrix(y_train, y_train_pred)
```

The confusion matrix shows that on the training data, of the 50 class 0 points, 46 were correctly classified as 0, and 4 incorrectly as 1; and of the 50 type 1 points, 2 were incorrectly classified as class 1 and 48 correctly as class 0. Another way of reporting the result would be to say that the predictions were 94% accurate. Note that since the data sets overlap, 100% accuracy is not possible.

(This all assumes that the data was generated using

`X_train,y_train = make_blobs(n_samples=100,centers=[(1,3),(2,0)],random_state=1)`)


```python
X_test,y_test = make_blobs(n_samples=100,centers=[(1,3),(2,0)],random_state=2)
y_hat_test  = output(w,b,X_test) 
y_test_pred = (y_hat_test >= 0.5).astype(int)
confusion_matrix(y_test, y_test_pred)
```

For this test data, the prediction is again 94% accurate.
